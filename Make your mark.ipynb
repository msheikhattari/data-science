```python
#Part 1
import folium
import requests
import pandas as pd
import math
import numpy as np

overdose_table = pd.read_csv("overdose.csv")

del overdose_table["Incident Date"]
del overdose_table["Year"]
del overdose_table["Opioid Overdose Review"]
del overdose_table["Treatment or Medication Given"]
del overdose_table["Incident Disposition"]
del overdose_table["Unit Name"]
del overdose_table["PD Narcan"]
del overdose_table["PD Save"]
del overdose_table["Complaint Reported by Dispatch"]
del overdose_table["Incident_ID"]
del overdose_table["Location"]

overdose_table = overdose_table.dropna(subset = ['Latitude', 'Longitude'])

print(overdose_table[pd.notnull(overdose_table["Latitude"])]["Latitude"].count())
overdose_table
```

The dataset used was opioid overdoses in Mesa, AZ since 2018. The dataset contains information on the year, date, unit, incident disposition, whether Narcan was used and also whether the patient was saved, the complaint reported, incident id, location, age, gender, as well as latitude and longitude data. Much of the data was superfluous so it was dropped, so that the only data kept was relating to the date of the incident, age, and gender of the patient. Then the rows with nan latitude/longitude data were dropped. The dataframe is shown along with the number of rows with non-nan location data in order to confirm all the operations were as intended. 

```python
#Part 2
map_osm = folium.Map(location=[33.42, -111.75], zoom_start=11.5)
map_osm
```

A map was generated centered on Mesa, AZ using the map function of the folium library as well as the coordinates from the dataset. 

```python
#Part 3
for index, row in overdose_table[overdose_table["Month"] == "Feb"].iterrows():
    if row["Gender"] == "Male":
        folium.Marker(location=[row["Latitude"], row["Longitude"]],
                    icon=folium.Icon(color='red')).add_to(map_osm) 
    if row["Gender"] == "Female":
        folium.Marker(location=[row["Latitude"], row["Longitude"]],
                    icon=folium.Icon(color='blue')).add_to(map_osm)
    
map_osm 
```

In the above map, the data was filtered so that only overdoses which occured in February were plotted. This was in order to reduce the amount of points on the map in order to better recognize trends in the data. Then the points were plotted on the basis of gender, so that male points were plotted with red markers, and female points plotted with blue markers. We can clearly observe that most of the points are clustered in western Mesa, with some of the points scattered to the East. Also, the overwhelming majority of the overdoses were men, evidenced by the disproportionate quantity of red markers relative to blue. The points are all relatively clustered towards the center in general, indicative that there are particular hotbeds of overdoses, whereas there are pockets on the map where there are no recorded overdoses. The points appear to reside mainly along both sides of the main highway which runs through Mesa. Overall this map helps with visualizing the demographic differences in overdoses as well as the distribution of them through the city.
